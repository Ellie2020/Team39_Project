```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

ladybirds_df1=df1[['Age', 'HoursLearning', 'Gender']]
ladybirds_df2=ladybirds_df1.dropna()
# Subset the fcc dataset by gender
male = ladybirds_df2.query("Gender == 'male'")
#female = ladybirds_df2.query("Gender == 'female'")
# Set up the figure
f, ax = plt.subplots(figsize=(10, 10))
ax.set_aspect("equal")
# Draw the two density plots
ax = sns.kdeplot(male.Age, male.HoursLearning,
                 cmap="Blues", shade=True, shade_lowest=False)
ax.tick_params(labelsize=16)
#ax = sns.kdeplot(female.HoursLearning, female.Age,
                 #cmap="Reds", shade=True, shade_lowest=False)
# Add labels to the plot
#red = sns.color_palette("Reds")[-2]
blue = sns.color_palette("Blues")[-2]
#ax.text(2.5, 8.2, "female", size=16, color=blue)
ax.text(3.8, 4.5, "male", size=22, color=blue)
plt.xlabel('Age (years)', fontsize=18)
plt.ylabel('Hours Learning (h/week)', fontsize=18)
plt.ylim((-5, 60))
plt.xlim((10, 60))
```
